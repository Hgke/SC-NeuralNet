```python
import numpy as np
from matplotlib import pyplot as plt
import sys
import os
import pickle
sys.path.insert(0, r'C:\Users\Admin\main\Science\Coding\SC-NeuralNet\solver')
from Objects import Synaps, JTL, Square_pulse, R, L
from FunctionCompiler import *
```

```python
Amplitude_steps = 0.1

Amplitude_signal_array = np.arange(0.5, 5, Amplitude_steps)  # амплитуда входного тока

dict_1_JJ1 = {}
dict_1_JJ10 = {}
dict_2_JJ1 = {}
dict_2_JJ10 = {}
count = 0
```

```python
for amplitude_signal in Amplitude_signal_array:
    
    jtl_1 = JTL.JTL(loc=[1, 2], N=10, ib_val=0.75, l_val=3)
    Synaps_1 = Synaps.Synaps(loc=[2, 3], L_value=1,R_value=0.1,C_value=1)
    r_12 = R.R(loc=[3, 4], r=0.1)
    jtl_2 = JTL.JTL(loc=[4, 5], N=10, ib_val=0.75, l_val=3)
    pulse = Square_pulse.Square_pulse(loc=[1], t0=50, A=amplitude_signal, length=100)

    t = np.linspace(0,400,1200)
    #t = np.arange(0, 400, 0.5)
    fc = FunctionCompiler([jtl_1,Synaps_1, r_12,jtl_2,pulse], t)
    sol = fc.solve()

    count += 1
    print(f'{count} из {len(Amplitude_signal_array)}')

    dict_1_JJ1[tuple([amplitude_signal])] = fc.object_dict['JTL1_JJ1'].get_data("V", t,sol)
    dict_1_JJ10[tuple([amplitude_signal])] = fc.object_dict['JTL1_JJ10'].get_data("V", t,sol)
    dict_2_JJ1[tuple([amplitude_signal])] = fc.object_dict['JTL2_JJ1'].get_data("V", t,sol)
    dict_2_JJ10[tuple([amplitude_signal])] = fc.object_dict['JTL2_JJ10'].get_data("V", t,sol)

    with open("data/dict_1_JJ1.pkl", "wb") as f1:
        pickle.dump(dict_1_JJ1, f1)
        f1.close()
        
    with open("data/dict_1_JJ10.pkl", "wb") as f2:
        pickle.dump(dict_1_JJ10, f2)
        f2.close()
        
    with open("data/dict_2_JJ1.pkl", "wb") as f3:
        pickle.dump(dict_2_JJ1, f3)
        f3.close()
        
    with open("data/dict_2_JJ10.pkl", "wb") as f4:
        pickle.dump(dict_2_JJ10, f4)
        f4.close()
```

```python
with open(r"data/dict_1_JJ1.pkl", "rb") as f:
    dict_1_JJ1 = pickle.load(f)
with open(r"data/dict_1_JJ10.pkl", "rb") as f:
    dict_1_JJ10 = pickle.load(f)
with open(r"data/dict_2_JJ1.pkl", "rb") as f:
    dict_2_JJ1 = pickle.load(f)
with open(r"data/dict_2_JJ10.pkl", "rb") as f:
    dict_2_JJ10 = pickle.load(f)
```

```python
# Keys
N = 1200
t = np.linspace(0, 400, N)

Amplitude_steps = 0.1
Amplitude_signal_array = np.arange(0.5, 5, Amplitude_steps)
print(Amplitude_signal_array)
```

```python
### JTL1_JJ10

#Analysis

for i in Amplitude_signal_array:
    
    y = dict_1_JJ10[(i,)]
    tstep = max(t)/N #signal freq

    Fs = 1/tstep #Sampling freq

    fstep = Fs/N
    f = np.linspace(0, (N-1)*fstep,N)

    #perform fft
    X = np.fft.fft(y)
    X_mag = np.abs(X)

    f_plot = f[0:int(N/2+1)]
    X_mag_plot_1_10 = 2*X_mag[0:int(N/2+1)]
    X_mag_plot_1_10[0] /= 2 

    fig, ax= plt.subplots(dpi=100)

    ax.set_xlabel('freq, Hz',fontsize=15)
    ax.plot(f_plot*2*np.pi,X_mag_plot_1_10)
    plt.title(f'JTL1_JJ10, A={np.round(i,1)}',fontsize=15)
    fig.savefig(f'graphics\JTL1_finish\Amplitude={np.round(i,1)}.jpg', bbox_inches='tight')
    plt.close("all")
    ### JTL2_JJ1

    #Analysis
    y = dict_2_JJ1[(i,)]
    tstep = max(t)/N #signal freq

    Fs = 1/tstep #Sampling freq

    fstep = Fs/N
    f = np.linspace(0, (N-1)*fstep,N)

    #perform fft
    X = np.fft.fft(y)
    X_mag = np.abs(X)

    f_plot = f[0:int(N/2+1)]
    X_mag_plot_2_1 = 2*X_mag[0:int(N/2+1)]
    X_mag_plot_2_1[0] /= 2 

    fig, ax= plt.subplots(dpi=100)

    ax.set_xlabel('freq, 10^12 Hz',fontsize=15)
    ax.plot(f_plot*2*np.pi,X_mag_plot_2_1)
    plt.title(f'JTL2_JJ1, A={np.round(i,1)}',fontsize=15)
    fig.savefig(f'graphics\JTL2_start\Amplitude={np.round(i,1)}.jpg', bbox_inches='tight')
    plt.close("all")
    # Transmission Function

    fig, ax= plt.subplots(dpi=100)

    ax.set_xlabel('freq, 10^12 Hz',fontsize=15)
    ax.plot(f_plot*2*np.pi,X_mag_plot_2_1/X_mag_plot_1_10)
    
    plt.title(f'Сравнение, A={np.round(i,1)}',fontsize=15)
    fig.savefig(f'graphics\Comparing_fourier\Amplitude={np.round(i,1)}.jpg', bbox_inches='tight')
    plt.close("all")
    
    fig, [ax1, ax2] = plt.subplots(nrows=2,ncols=1,dpi=100)
    ax1.plot(t,dict_1_JJ10[(i,)])
    ax1.set_ylabel('№1_JJ10')
    ax2.set_ylabel('№2_JJ1')
    ax2.plot(t,dict_2_JJ1[(i,)])
    plt.xlabel(f'Сравнение, A={np.round(i,1)}',fontsize=15)
    fig.savefig(f'graphics\Comparing_trivial\Ampl={np.round(i,1)}.jpg', bbox_inches='tight')
    plt.close("all")

```
